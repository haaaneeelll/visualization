```python
####################################################
### 3. 선그래프
'''
   선그래프
   - plot() 함수는 선(line) 또는 마커(marker)를 그릴 때 사용.
   - 하나의 리스트는 y값으로 인식.
   - xlabel() 과 ylabel() 함수는 x축과 y축의 레이블을 지정한다.
    
   - 축 범위 지정
       plt.axis([0, 10, 0, 20]) # [xmin, xmax, ymin, ymax] 형태로 X, Y축의 범위를 지정한다.
       입력값이 없으면 데이터에 맞게 자동으로 범위를 지정한다.
    
   - 눈금 지정
     plt.xticks(range(11)) # x축 눈금지정
     plt.yticks([0,1,5,10,20]) # y축 눈금지정

   - 눈금 지정 + label
     plt.xticks([1,2,3,4], labels=['Jan',  'Feb',  'Mar',  'May'])
     plt.yticks([1, 4, 9, 16], ('0', '1GB', '2GB', '3GB'))

   - 색상지정
     plt.plot([1,2,3,4],[1, 4, 9, 16], color="violet")
     https://matplotlib.org/stable/tutorials/colors/colors.html#specifying-colors
    
   - marker 지정
    plt.plot([1,2,3,4],[1, 4, 9, 16], "P") #  o, s, p,..
    plt.plot([1,2,3,4],[1, 4, 9, 16], marker="P")
     https://matplotlib.org/stable/api/markers_api.html?highlight=marker#module-matplotlib.markers

   - 선 스타일
   plt.plot([1,2,3,4],[1, 4, 9, 16], marker="P", linestyle="--")
   # https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_linestyle
'''

# 1) y값만 지정
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8,6))  # 생략가능

plt.plot([1,2,3,4])  # y 값 , x 값은 자동 생성
plt.show()
```

```python
# 2) x,y값 지정
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8,6))  # 생략가능
plt.plot([1,2,3,4],[1, 4, 9, 16])  # x 값 , y 값
plt.show()
```

```python
# 3) x,y 라벨지정 
import matplotlib.pyplot as plt
plt.rc('font', family='Helvetica')
plt.rc('font', size=15)

fig = plt.figure(figsize=(8,6))  # 생략가능
plt.plot([1,2,3,4],[1, 4, 9, 16])  # x 값 , y 값
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.show()
```

```python
# 4) 축 범위 지정
import matplotlib.pyplot as plt

# fig = plt.figure(figsize=(8,6))  # 생략가능
plt.plot([1,2,3,4],[1, 4, 9, 16])  # x 값 , y 값
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.axis([0, 5, 0, 20]) # [xmin, xmax, ymin, ymax] 형태로 X, Y축의 범위를 지정한다.
print(plt.axis())  # (0.0, 5.0, 0.0, 20.0)
plt.show() 
```

```python
# 5) 눈금 지정
# fig = plt.figure(figsize=(8,6))  # 생략가능
plt.plot([1,2,3,4],[1, 4, 9, 16])  # x 값 , y 값
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.axis([0, 10, 0, 20]) # [xmin, xmax, ymin, ymax] 형태로 X, Y축의 범위를 지정한다.
plt.xticks([0,5,10]) # x축 눈금지정
# plt.xticks(range(11)) # x축 눈금지정
plt.yticks([0,1,5,10,20]) # y축 눈금지정
plt.show()
```

```python
# 6) 눈금 지정 + label
import matplotlib.pyplot as plt
import numpy as np

# fig = plt.figure(figsize=(8,6))  # 생략가능
plt.plot([1,2,3,4],[1, 4, 9, 16])  # x 값 , y 값
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
# plt.xticks([1,2,3,4], labels=['Jan',  'Feb',  'Mar',  'May'])
plt.xticks([1,2,3,4], labels=['Jan',  'Feb',  'Mar',  'May'], rotation=45)
plt.yticks([1, 4, 9, 16], ('0', '1GB', '2GB', '3GB'))
plt.show()
```

```python
# 7)색상지정
# https://matplotlib.org/stable/tutorials/colors/colors.html#specifying-colors
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[1, 4, 9, 16], color="violet")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.axis([0, 5, 0, 20])
plt.show()
```

```python
# 8)marker 지정
# https://matplotlib.org/stable/api/markers_api.html?highlight=marker#module-matplotlib.markers

import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[1, 4, 9, 16], "*") #  o, s, p, ..
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.axis([0, 5, 0, 20])
plt.show()
```

```python
# 9) 선 스타일
# https://matplotlib.org/stable/api/markers_api.html?highlight=marker#module-matplotlib.markers

import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[1, 4, 9, 16], linestyle = "--") #  o, s, p, ..
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.axis([0, 5, 0, 20])
plt.show()
```

```python
####################################################
### 4. 여러 그래프 한번에 그리기
'''
	범례(legend)

	1. plt.plot(.., label="라벨1")
	   plt.plot(.., label="라벨2")
	   plt.legend()

	2.
	   plt.plot(..)
	   plt.plot(..)
	   plt.legend(["라벨1", "라벨2" ])
'''
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(0,2, 0.2)

plt.plot(n,n, 'bo', label = 'linear')
plt.plot(n,n**2, 'r', marker="*", linewidth=2, label = 'quadratic')
plt.plot(n,n**3, 'y', marker="^", linewidth=3, label = 'cubic')
plt.legend(["linear","quadratic","cubic"])
plt.show()
```

```python
####################################################
### 4. 그리드
'''
	범례(legend)

	1. plt.plot(.., label="라벨1")
	   plt.plot(.., label="라벨2")
	   plt.legend()

	2.
	   plt.plot(..)
	   plt.plot(..)
	   plt.legend(["라벨1", "라벨2" ])
'''
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(0,2, 0.5)
plt.grid(True, axis = 'both', color = "blue", linestyle = "--")
plt.plot(n,n, 'bo', label = 'linear')
plt.plot(n,n**2, 'r', marker="*", linewidth=2, label = 'quadratic')
plt.plot(n,n**3, 'y', marker="^", linewidth=3, label = 'cubic')
plt.legend(["linear","quadratic","cubic"])
plt.show()
```
