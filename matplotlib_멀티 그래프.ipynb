```python
####################################################
### 1. 기본

import matplotlib.pyplot as plt
# 1. 시각화하기 위한 도화지 생성
fig = plt.figure(figsize=(8,6))  # figsize=( w, h ) 단위는 inch
# 2. 도화지에 그리기 위한 붓 생성 ==> 거의 사용 안하고 plt로 직접 그래프 그림
ax = plt.axes()
plt.show()
```

```python
####################################################
### 2. 다중 그래프

# 가) fig.add_subplot(행,열, n번째열)
import matplotlib.pyplot as plt
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)  # 1행 2열의 1번째열
plt.plot([1,2,3,4])  # ax1.plot([1,2,3,4]) 대신에 plt 사용
plt.title("sub1")

ax2= fig.add_subplot(1,2,2)   # 1행 2열의 2번째열
plt.plot([-1,-2,-3,-4])
plt.title("sub2")

fig.suptitle("Main")
plt.show()
```

```python
# 나) plt.subplots(행,열)
import matplotlib.pyplot as plt
fig, axs = plt.subplots(nrows=1, ncols=2) # 1행 2열
print(axs[0])
axs[0].plot([1,2,3])
axs[1].plot([-1,-2,-3])
plt.show()
```

```python
# 나) plt.subplots(행,열)
fig, axs = plt.subplots(nrows=1, ncols=2) # 1행 2열
print(axs)
axs[0].plot([1,2,3])
axs[1].plot([-1,-2,-3])
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows=2, ncols=2)  # 1행 2열
print(axs)
axs[0, 0].plot([1, 2, 3])  # 첫 번째 행, 첫 번째 열의 서브플롯에 그래프 그리기
axs[0, 1].plot([-1, -2, -3])  # 첫 번째 행, 두 번째 열의 서브플롯에 그래프 그리기
axs[1, 0].plot([4, 5, 6])  # 두 번째 행, 첫 번째 열의 서브플롯에 그래프 그리기
axs[1, 1].plot([-4, -5, -6])  # 두 번째 행, 두 번째 열의 서브플롯에 그래프 그리기

plt.show()

```
